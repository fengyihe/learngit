```python
df3 = df2.drop(['媒体类型','注册新账号设备数','首日充值'],axis =1)
df3.head()
```

```python
df4 = df3.loc[(df3["广告花费"] > 0) | ((df3["激活设备数"] > 0) & (df3["注册设备数"] > 0 ))]
# 筛选出来的数据是包含广告花费不为0的行，以及激活设备数和注册设备数同时大于0的行
```

```python
df5 = df4.sort_values(by=["广告花费","注册设备数","激活设备数"] , ascending=False)
# 现在的问题就是如何删除花费为0，也不是残留的量
# 当列数及列名一致的时候，这里就涉及到DataFrame的操作。Pandas中DataFrame数据合并、连接（concat、merge、join）
```

```python
df5.to_excel('C:\\Users\\qinzc\\Desktop\\20181109.xlsx',index=False) # 设置index=False不导入索引
# 保存完之后删除原下载文件
```

```python
#载入已有工作表
from openpyxl.workbook import Workbook
from openpyxl.writer.excel import ExcelWriter
from openpyxl import load_workbook
import datetime

# C:\\Users\\qinzc\\Desktop\\20181108\\问道日报20181108\\《问道手游》日报数据(11.8).xlsx
```

```python
#载入已有工作表
from openpyxl.workbook import Workbook
from openpyxl.writer.excel import ExcelWriter
from openpyxl import load_workbook
# 读取已有工作簿
wb = load_workbook(filename = 'test_write.xlsx') #这里是工作日报
# 针对工作簿创建新的工作表，并用ws2来表示
ws2 = wb.create_sheet(title="data") # 对读入的工作表创建新工作表
for i in range(10):
    ws2.cell(row=1,column=1+i).value= i
# 其中对新建的工作表进行变更之后，对工作簿ws进行保存即可。
wb.save('test_write.xlsx')
```
